```python
__author__ = 'Алексей Сидорюк'

'''
Примечание:
1. Постарайтесь решить задачи без использования массивов. Им будет посвящён следующий урок.
'''

# 1. Написать программу, которая будет складывать, вычитать, умножать или делить два числа.
# Числа и знак операции вводятся пользователем. После выполнения вычисления программа не должна завершаться,
# а должна запрашивать новые данные для вычислений. Завершение программы должно выполняться при вводе символа '0'
# в качестве знака операции. Если пользователь вводит неверный знак (не '0', '+', '-', '*', '/'),
# то программа должна сообщать ему об ошибке и снова запрашивать знак операции.
# Также сообщать пользователю о невозможности деления на ноль, если он ввел 0 в качестве делителя.

def calc():
    num1 = input('Введите первое число: ')
    oper = input('Введите операцию: ')
    num2 = input('Введите второе число: ')
    if num1.isdigit() and num1.isdigit(): # пока будем работать только с целыми числами
        if oper == '0':
            return
        elif oper == '+':
            print('{} + {} = {}'.format(num1, num2, int(num1) + int(num2)))
            return calc()
        elif oper == '-':
            print('{} - {} = {}'.format(num1, num2, int(num1) - int(num2)))
            return calc()
        elif oper == '*':
            print('{} * {} = {}'.format(num1, num2, int(num1) * int(num2)))
            return calc()
        elif oper == '/':
            if int(num2) != 0:
                print('{} / {} = {}'.format(num1, num2, int(num1) / int(num2)))
            else:
                print('Деление на 0.')
            return calc()
        else:
            print('Неправильный знак операции.')
            return calc()
    else:
        return calc()

calc()
```

```python
# 2. Посчитать четные и нечетные цифры введенного натурального числа.
# Например, если введено число 34560, то у него 3 четные цифры (4, 6 и 0) и 2 нечетные (3 и 5).

even = 0
odd = 0
num = input()

if num.isnumeric():
    for i in num:
        if i.isnumeric():
            if int(i)%2:
                odd += 1
            else:
                even += 1
    print ('Введенное число содержит {} четных и {} нечетных цифр.'.format(even, odd))
else:
    print('Введено не целое число.')
```

```python
# 3. Сформировать из введенного числа обратное по порядку входящих в него цифр и вывести на экран.
# Например, если введено число 3486, то надо вывести число 6843.
# О том, что строки - объект итерируемый, забыли. Реверс рекурсивный.

def reverse_num(num_str, num_position):
    if int(num_str) >= 0:
        # базовый случай - дошли до конца строки (разряд единиц в представлении целого цисла)
        if num_position == 1:
            return str(int(num_str) % 10**num_position // 10**(num_position - 1))
        else:
            return reverse_num(num_str, num_position - 1) + str(int(num_str) % 10**num_position // 10**(num_position - 1))
    else:
        return '-' + reverse_num(str(abs(int(num_str))), num_position - 2) + str(abs(int(num_str)) % 10**(num_position - 1) // 10**(num_position - 2))

num_str = input('Введите целое число (последовательность цифр):')

print('Обратная последовательность: ' + reverse_num(num_str, len(num_str)))
```

```python
# 4. Найти сумму n элементов следующего ряда чисел: 1 -0.5 0.25 -0.125 ...Количество элементов (n) вводится с клавиатуры.

def sum_series(n):
    if n == 0:
        return (-1/2)**n
    elif n > 0:
        return (-1/2)**n + sum_series(n - 1)

n = input('Введите количество элементов ряда: ')
print(sum_series(int(n)))
```

```python
# 5. Вывести на экран коды и символы таблицы ASCII, начиная с символа под номером 32 и заканчивая 127-м включительно.
# Вывод выполнить в табличной форме: по десять пар "код-символ" в каждой строке.

row = ''
for i in range(32,128):
    row += '"' + str(i) + '-' + chr(i) + '"  '
    if (i-1)%10 == 0:
        print(row, end='\n')
        row = ''
print(row)
```
